```python
#!/usr/bin env python
import pandas as pd


def func(df):

    df_plot = df.plot(kind='bar', figsize=(20, 16), fontsize=18, stacked=True, rot=0, width=0.3, legend=False)
    df_plot.legend(fontsize=20)
    fig = df_plot.get_figure()
    fig.savefig('/home/sa.goyal/Desktop/test.png')

    
def main():
    dt_rng = pd.date_range(start='2017-01-01', periods=5, freq='B')
    df = pd.DataFrame({
        'P' : [1,2,3,4,5],
        'F' : [6,7,8,9,10],
        'I' : [10,12,13,14,15]
    }, columns=['P','F','I'], index=dt_rng)
    func(df)

if __name__ == '__main__':
    main()
```
