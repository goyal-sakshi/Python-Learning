```python
# Linked List

class Node:
    
    def __init__(self, data):
        self.data = data
        self.next = None
    
class LinkedList:
    
    def __init__(self):
        self.head = None
        
    def insert(self, newNode):
        
        if self.head is None:
            self.head = newNode
        else:
            currentNode = self.head
            while True:
                if currentNode.next is None:
                    break
                currentNode = currentNode.next
            currentNode.next = newNode
                    
    def printList(self):
        currentNode = self.head
        print "Linked List:"
        while True:
            if currentNode is None:
                break
            print currentNode.data
            currentNode = currentNode.next
            
    def get_node_at_pos(self, position):
        currentNode = self.head
        count = 1
        while True:
            if currentNode.next is None:
                break
            currentNode = currentNode.next
            count += 1
            if count == position:
                return currentNode.data
            
    def insert_at_pos(self, newNodeData, position):
        currentNode = self.head
        count = 1
        while True:
            if currentNode.next is None:
                break
            currentNode = currentNode.next
            if count + 1 == position:
                n = Node(newNodeData)
                n.next = currentNode.next
                currentNode.next = n
                print "*****",n.data
            count += 1

n1 = Node(1)
n2 = Node(2)
n3 = Node(3)

# head=> 1-> 2-> 3-> None
ll = LinkedList()
ll.insert(n1)
ll.insert(n2)
ll.insert(n3)
ll.insert(Node(4))

ll.printList()

print "Element at position 2 is :", ll.get_node_at_pos(2)
# arg1 => data
# arg2 => position
ll.insert_at_pos(9, 2)
ll.printList()
```
